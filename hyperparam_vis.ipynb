```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
learning_rates = [0.0001, 0.0005, 0.001, 0.005]
train_loss = [1.7647, 1.9388, 2.0638, 2.4140]
val_loss = [1.7825, 1.9301, 2.0398, 2.3759]
train_mse = [1.7647, 1.9388, 2.0638, 2.4140]
val_mse = [1.7825, 1.9301, 2.0398, 2.3759]
train_mae = [1.0615, 1.1323, 1.1768, 1.2792]
val_mae = [1.0705, 1.1323, 1.1736, 1.2733]
train_r2 = [0.2493, 0.1752, 0.1248, -0.0166]
val_r2 = [0.2312, 0.1685, 0.1222, -0.0171]

# Create the figure
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(10, 10))

# Plot the loss and R2 values
ax1_1 = ax1.twinx()
ax1.plot(learning_rates, train_loss, label='Train Loss')
ax1.plot(learning_rates, val_loss, label='Val Loss')
ax1_1.plot(learning_rates, train_r2, '--', label='Train R2')
ax1_1.plot(learning_rates, val_r2, '--', label='Val R2')
ax1.set_xlabel('Learning Rate')
ax1.set_ylabel('Loss')
ax1_1.set_ylabel('R2')
ax1.set_title('Loss and R2 vs Learning Rate')
ax1.legend(loc='upper left')
ax1_1.legend(loc='upper right')

# Plot the MSE and MAE values
ax2_1 = ax2.twinx()
ax2.plot(learning_rates, train_mse, label='Train MSE')
ax2.plot(learning_rates, val_mse, label='Val MSE')
ax2_1.plot(learning_rates, train_mae, '--', label='Train MAE')
ax2_1.plot(learning_rates, val_mae, '--', label='Val MAE')
ax2.set_xlabel('Learning Rate')
ax2.set_ylabel('MSE')
ax2_1.set_ylabel('MAE')
ax2.set_title('MSE and MAE vs Learning Rate')
ax2.legend(loc='upper left')
ax2_1.legend(loc='upper right')


# Define the data
batch_sizes = [128, 256, 512]
train_loss = [1.7855, 1.8235, 1.8725]
val_loss = [1.7990, 1.8379, 1.8749]
train_mse = [1.7855, 1.8235, 1.8725]
val_mse = [1.7990, 1.8379, 1.8749]
train_mae = [1.0664, 1.0796, 1.0944]
val_mae = [1.0746, 1.0875, 1.0994]
train_r2 = [0.2513, 0.2403, 0.2216]
val_r2 = [0.2333, 0.2236, 0.2081]


# Plot the loss and R2 values
ax3_1 = ax3.twinx()
ax3.plot(batch_sizes, train_loss, label='Train Loss')
ax3.plot(batch_sizes, val_loss, label='Val Loss')
ax3_1.plot(batch_sizes, train_r2, '--', label='Train R2')
ax3_1.plot(batch_sizes, val_r2, '--', label='Val R2')
ax3.set_xlabel('Batch Size')
ax3.set_ylabel('Loss')
ax3_1.set_ylabel('R2')
ax3.set_title('Loss and R2 vs Batch Size')
ax3.legend(loc='upper left')
ax3_1.legend(loc='upper right')

# Plot the MSE and MAE values
ax4_1 = ax4.twinx()
ax4.plot(batch_sizes, train_mse, label='Train MSE')
ax4.plot(batch_sizes, val_mse, label='Val MSE')
ax4_1.plot(batch_sizes, train_mae, '--', label='Train MAE')
ax4_1.plot(batch_sizes, val_mae, '--', label='Val MAE')
ax4.set_xlabel('Batch Size')
ax4.set_ylabel('MSE')
ax4_1.set_ylabel('MAE')
ax4.set_title('MSE and MAE vs Batch Size')
ax4.legend(loc='upper left')
ax4_1.legend(loc='upper right')

# Define the data
epochs = [100, 200, 500]
train_loss = [1.7561, 1.6827, 1.5496]
val_loss = [1.7676, 1.7247, 1.6338]
train_mse = [1.7561, 1.6827, 1.5496]
val_mse = [1.7676, 1.7247, 1.6338]
train_mae = [1.0554, 1.0348, 0.9866]
val_mae = [1.0620, 1.0505, 1.0116]
train_r2 = [0.2543, 0.2843, 0.3378]
val_r2 = [0.2381, 0.2551, 0.2933]


# Plot the loss and R2 values
ax5_1 = ax5.twinx()
ax5.plot(epochs, train_loss, label='Train Loss')
ax5.plot(epochs, val_loss, label='Val Loss')
ax5_1.plot(epochs, train_r2, '--', label='Train R2')
ax5_1.plot(epochs, val_r2, '--', label='Val R2')
ax5.set_xlabel('Epochs')
ax5.set_ylabel('Loss')
ax5_1.set_ylabel('R2')
ax5.set_title('Loss and R2 vs Epochs')
ax5.legend(loc='upper left')
ax5_1.legend(loc='upper right')

# Plot the MSE and MAE values
ax6_1 = ax6.twinx()
ax6.plot(epochs, train_mse, label='Train MSE')
ax6.plot(epochs, val_mse, label='Val MSE')
ax6_1.plot(epochs, train_mae, '--', label='Train MAE')
ax6_1.plot(epochs, val_mae, '--', label='Val MAE')
ax6.set_xlabel('Epochs')
ax6.set_ylabel('MSE')
ax6_1.set_ylabel('MAE')
ax6.set_title('MSE and MAE vs Epochs')
ax6.legend(loc='upper left')
ax6_1.legend(loc='upper right')



# Adjust the spacing between subplots
plt.subplots_adjust(hspace=0.5)
plt.tight_layout()
# Display the plot
plt.show()
```
